## Average per country
Calculate the average for each country, put it into a function that take a dataframe as input and print the results

*Hint*:
1. Create a list that will have the country name (*Hint*: you can use the ```unique()``` method)
1. Use a for loop over all the element of this list
1. Inside the loop create a subset dataframe for the given country
1. Calculate the average (using ```mean()``` method) of this subset for the correct columns (you might have to do ```mean()``` two times)
1. Print the average with the country name

````python
# SOLUTION:
def average_percountry(dataframe):
    list_country=dataframe["country"].unique()
    for country in list_country:
        subset=dataframe[dataframe['country']==country]
        mean=subset.loc[:, "2011 mean":"2017 mean"].mean()
        mean=mean.mean()
        print(country, ':',mean)
average_percountry(amphetamine_ori)   
````

### Average per country output

Based on the previous function, we would like to return a dataframe with the average values per country and the country name as index.

To create a dataframe you can use : ```pd.DataFrame(_List_, index=_List_,columns=_List_)```.

*Hint*: Put the result of each country on a list then create the dataframe after the loop.

**Extra 1**: Remove country that have an average lower than or equal to zero.

**Extra 2**: add the averaged standard deviation to the dataframe

* You then can plot the result in a bar graph
* You can also save this dataframe using ```to_csv()``` method [See Doc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html?highlight=to_csv#pandas.DataFrame.to_csv)

````python
# SOLUTION:
def average_percountry(dataframe):
    list_country=dataframe["country"].unique()
    outlist=[]
    for country in list_country:
        subset=dataframe[dataframe['country']==country]
        mean=subset.loc[:, "2011 mean":"2017 mean"].mean().mean()

        outlist.append(mean)
    df=pd.DataFrame(outlist, index=list_country,columns=["average",])
    return df
average_percountry(amphetamine_ori).plot(kind='bar',figsize=(12, 12))
````

### With error bar and removed null average

````python
# SOLUTION:
def average_percountry(dataframe):
    list_country=dataframe["country"].unique()
    outlist=[]
    for country in list_country:
        subset=dataframe[dataframe['country']==country]
        mean=subset.loc[:, "2011 mean":"2017 mean"].mean().mean()
        std=subset.loc[:, "2011 sd":"2017 sd"].mean().mean()

        outlist.append([mean,std])
    df=pd.DataFrame(outlist, index=list_country,columns=["average","std"])
    return df[df["average"]>0]
var=average_percountry(amphetamine_ori)
var['average'].plot(kind='bar',figsize=(12, 12),yerr=var['std'],capsize=4)
````

### Make a csv file

````python
# SOLUTION:
var.to_csv('amphetamine_mean_percountry.csv')
````

## Make a script
Make a script that:
1. open a csv file
1. Calculate the average of each country (function developped before)
1. save this average into another csv file with the pattern "nameoffile_meanpercountry.csv

````python
# SOLUTION:
import pandas as pd

def average_percountry(dataframe):
    list_country=dataframe["country"].unique()
    outlist=[]
    for country in list_country:
        subset=dataframe[dataframe['country']==country]
        mean=subset.loc[:, "2011 mean":"2017 mean"].mean().mean()
        std=subset.loc[:, "2011 sd":"2017 sd"].mean().mean()

        outlist.append([mean,std])
    df=pd.DataFrame(outlist, index=list_country,columns=["average","std"])
    return df[df["average"]>0]
file="./data/score_benzoylecgonine_wastewater_loads_2011_2017"
data=pd.read_csv(file+".csv",delimiter=";",index_col=0)
avg_df=average_percountry(data)
avg_df.to_csv(file+"_meanpercountry.csv")
````
